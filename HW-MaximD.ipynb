```python
                                     ВСТРОЕННЫЕ ОПЕРАЦИИ

1. abs() - находит модуль числа и принимает на вход одно значение. 
Модуль числа - это абсолютная величина, то есть, грубо говоря, модуль числа отбрасывает знак.
( то есть abs(-7) => 7, abs(9) => это 9, так же это работает с рациональными числами. Здесь же
 можно выполнять различные вычисления, НАПРИМЕР abs(1-100) => 99)

2. min() и max() - min принимает на вход несколько значений через запятую и находит самое наименьшее из них.
max - в свою очередь, найдет самое наибольшее.
(min(1,2,4,6,11,-4,-9,-99) самым меньшим будет -99, это и будет нашим ответом)
(max(1,2,4,6,11,-4,-9,-99) для операции max это будет 11)

3. pow() - принимает на вход 2 значения и возводит первое число в степень второго.
(pow(2,4) => 16, это 2 в 4 степени)

4. round() - принимает на вход значение и выполняет округление по умолчанию до целого числа.
Также, если через запятую указать разряд, то функция будет округлять именно до этого разряда. 
(round(2.47) => 2, выдаст целое число, тогла как round(2.47,1) => 2.4, а round(2.47,2) => 2.47)

Также в round() можно передать как положительные так и отрицательные значения. 
Положительные значения отвечают за разряды, расположенные после точки в дробной части 
(например, 1=десятые, 2=сотые, 3=тысячные и т.д.). 
Отрицательными обозначаются разряды, расположенные до точки (-1=десятки, -2=сотни, -3=тысячи и т.д.) 
К примеру, если передать функции round() вторым значением -1, то она округлит ее до первого знака до запятой 
(до десятков)
(round(777,-1) => 770, округление до дестков, при значении второго аргумента -2, получится 800)

5. В Python есть функция type(), которая принимает объект и возвращает тип данного объекта.
НАПРИМЕР (type(4) выдаст такой результат <class 'int'>, а на type(4.42) <class 'float'>)

ВАЖНО ЗНАТЬ, что Внутри функций вы можете использовать и другие функции.
В таких случаях все выражения будут выполняться согласно приоритету операций.
Сначала выражения в скобках, и лишь затем результат самой функции.

                                     ФУНКЦИЯ input()

После вызова функции input() программа останавливает свое выполнение и ждет, когда пользователь введет текст. 
По окончанию ввода пользователь нажимает Enter , функция input() заберет введенный текст и 
передаст его программе, которая уже будет обрабатывать его согласно своим алгоритмам.

Функция input() позволяет сохранять данные или работать с ними в удобном формате, 
для этого функцию нужно обернуть в еще одну, НАПРИМЕР:

d = int(input()) или d = float(input())

Так же, функция input() позволяет вводить 2 значения с помощью .split():

a, b = input().split() # вводится ТОЛЬКО 2 числа через пробел
a = int(a)
b = int(b)
print(a, b)
print(type(a),type(b))

Три значения с помощью map(): 

a, b, c = map(int, input().split()) #считываем 3 целых числа через пробел
print(a,b,c)

Варианты использования функции input()

a = input() - если необходимо ввести строку и сохранить ее в переменную а
a = int(input()) - если необходимо ввести целое число и сохранить его в переменную а
a = float(input()) - если необходимо ввести вещественное число и сохранить его в переменную а
a,b = map(int,input().split()) - если необходимо ввести два целых числа в одну строку через пробел
a,b,c = map(float,input().split()) - если необходимо ввести три вещественных числа в одну строку через пробел

                                     Необязательные аргументы print()

1. Аргумент sep называется разделителем (от англ. «separator»). Он отвечает за то, 
что будет выводиться между значениями в пределах одной команды print. 
Но это значение можно изменить при помощи параметра sep. 
Ему нужно присвоить внутри функции print() новое значение в виде строки. 
НАПРИМЕР:

print(1, 2, 3, 4, sep='###') => 1###2###3###4

2. Аргумент end. От этого аргумента будет зависеть, что будет печататься после вывода всех значений 
в пределах одной команды print(). Команда print() по умолчанию  в конце производит перенос на новую строчку, 
потому что по умолчанию аргумент end принимает значение \n - это служебный символ, обозначающий перенос строки.

То есть, после команды print(), по умаолчанию происходит перенос на новую строку.
Потому, для отмены переноса, можно указать аргументу end другое значение, НАПРИМЕР: 

print(1, 2, 3, end='!') #не будет переноса в конце => 1 2 3!07.12.1996
print(07.12.1996)

                                     Деление нацело

В python данная операция обозначается знаком двойного слеша //. И допустим если вам надо нацело поделить a на b (a//b), 
нужно ответить на вопрос: «Cколько раз второе число (в нашем случае b) умещается в первое?»
В одном a может быть сколько угодно b.
Если число b = 100 & a = 10912, то ответ будет  109.

                                     Остаток от деления
                                     
Остаток от деления в python обозначается знаком процента %. Чтобы посчитать, вы должны ответить на вопрос:
«Сколько останется от первого числа после того, как в него максимальное количество раз уместится второе число?»
(a % b) если взять A = 14352 & B = 1000, то B поместится в A ровно 14 раз, а значит остатком будет число 352! и т д.
Если бы в данном случае B было бы равно 100000, то A было бы равно самому себе. 

                                     Составное присваивание

                                     Оператор 	Описание 	Синтаксис
+= 	Сложение с присвоением: складывает левый операнд с правым,
 и затем присваивает полученное значение левому операнду 	a += b

-= 	Вычитание с присвоением: вычитаем из левого операнда правый, 
и затем присваивает полученное значение левому операнду a -= b

*= 	Умножение с присвоением: перемножает левый операнд с правым, 
и затем присваивает полученное значение левому операнду 	a *= b

/= 	Деление с присвоением: делит левый операнд на правый, 
и затем присваивает полученное значение левому операнду 	a /= b

//= Деление нацело с присвоением: нацело делит левый операнд на правый, 
и затем присваивает полученное значение левому операнду 	a //= b

%= 	Деление по остатку присвоением: делит по остатку левый операнд на правый, 
и затем присваивает полученное значение левому операнду 	a %= b

**= Возведение в степень с присвоением: возводит левый операнд в степень правого операнда, 
и затем присваивает полученное значение левому операнду 	a **= b

                                     Булиевый тип операций


    == Возвращает True, если оба операнда равны. 
    != Возвращает True, если оба операнда НЕ равны. 
    > (больше чем) Возвращает True, если первый операнд больше второго.
    < (меньше чем) 
    >= (больше или равно) 
    <= (меньше или равно)


Функция trunc выполняет отсечение дробной части и оставляет только целую часть числа.
Функция floor выполняет округление вниз до ближайшего целого числа.
Функция ceil выполняет округление вверх до ближайшего целого числа.
    
При делении -53 // 5 нацело отрицательного числа, необходимо понимать, что существует какое то расстояние 
или существуют числа от 0 до 5 как и от 0 до -53, в таком случае нужно учитывать, что от 5 до - 53
будет 58 пунктов, значит и ответом будет НЕ -10, А -11!

1. Деление положительного числа на положительное.

Делитель положительный, значит остаток положительный.
Допустим, 25 % 11. Делим 25 на 11, получаем 2 с чем-то, округляем к меньшему, 
получаем частное = 2. 25 - 11 * 2 = 3, остаток = 3. Тут все как обычно.

2. Деление положительного числа на отрицательное.

Делитель отрицательный, значит остаток отрицательный.
Допустим, 31 % (-5). Делим 31 на -5, получаем -6 с чем-то, округляем к меньшему 
(к минус бесконечности, еще раз уточню), получаем частное = -7. 31 - (-5) * (-7) = -4, 
остаток -4 (делитель отрицательный, значит остаток отрицательный, норм).

3. Деление отрицательного на положительное.

Делитель положительный, значит остаток положительный.
Допустим, -52 % 7. Делим -52 на 7, получаем -7 с чем-то, округляем к меньшему, 
получаем частное -8. -52 - 7 * (-8) = 4, остаток 4.

4. Деление отрицательного на отрицательное.

Делитель отрицательный, значит остаток отрицательный.
Допустим, -35 % (-14). Делим -35 на -14, получаем 2 с чем-то, округляем к меньшему, 
получаем частное 2. -35 - (-14) * 2 = -7, остаток -7.

                                     Операции над строками

Операция in проверяет есть ли какая-либо подстрока в нашей строке. 
Например, в нашей переменной s сейчас содержится слово Hello. Мы можем проверить содержатся ли 
в ней символ l или подстрока He.
Обратите внимание, что порядок символов и регистр важен. Результатом операции будет булево значение:

s 
'Hello'
'He' in s
True
'll' in s
True
'o' in s
True

Буквы можно так же сравнивать, для этого мы узнаем 
числовое значение каждого символа с помощью команды ord().
При этом больше будет та строка которая длиннее.

```
